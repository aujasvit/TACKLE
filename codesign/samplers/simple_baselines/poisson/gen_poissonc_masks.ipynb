```python
import numpy as np
import matplotlib.pyplot as plt
from sigpy.mri import poisson

def get_poisson_mask(shape, acc_ratio, seed, tol):
    pre_ratio = acc_ratio*8
    H, W = shape[0], shape[1]
    H_calib, W_calib = int(H / np.sqrt(pre_ratio)), int(W / np.sqrt(pre_ratio)) 
    mask = poisson(shape, acc_ratio, calib=(H_calib, W_calib), dtype=int, crop_corner=False, seed=seed, tol=tol)
    plt.figure()
    plt.imshow(mask, 'gray')
    plt.title(f'actual acceleration: {H*W/np.sum(mask):.2f}')
    plt.show()
    with open(f'shape={shape}_{acc_ratio}x.npy', 'wb') as f:
        np.save(f, mask)
```

## fastMRI+

```python
shape = (192, 192)

# 8x
get_poisson_mask(shape=shape, acc_ratio=8, seed=2, tol=0.01)

# 16x
get_poisson_mask(shape=shape, acc_ratio=16, seed=2, tol=0.1)
```

## OASIS 2D

```python
shape = (192, 160)

# 8x
get_poisson_mask(shape=shape, acc_ratio=8, seed=1, tol=0.01)

# 16x
get_poisson_mask(shape=shape, acc_ratio=16, seed=2, tol=0.01)

# 64x
get_poisson_mask(shape=shape, acc_ratio=64, seed=1, tol=0.01)
```
